```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del modelo
P0 = 200  # Tamaño inicial de la población
r = 0.08  # Tasa de crecimiento (8% por unidad de tiempo)
dt = 0.1  # Paso de tiempo (0.1 unidades de tiempo)
t_max = 20  # Tiempo total de la simulación (20 unidades de tiempo)

# Listas para almacenar el tiempo y la población
t = np.arange(0, t_max + dt, dt)
P = np.zeros(len(t))
P[0] = P0

# Matriz para almacenar el tiempo y la población
datos = np.zeros((len(t), 2))
datos[0, 0] = t[0]
datos[0, 1] = P0

# Método de Euler para simular el crecimiento de la población
for i in range(1, len(t)):
    P[i] = P[i-1] + r * P[i-1] * dt
    datos[i, 0] = t[i]
    datos[i, 1] = P[i]

# Graficar los resultados
plt.plot(t, P, label='Población')
plt.xlabel('Tiempo (unidades)')
plt.ylabel('Población')
plt.title('Simulación del Crecimiento de la Población con el Método de Euler')
plt.legend()
plt.grid(True)
plt.show()

# Tamaño final de la población
tamaño_final = P[-1]
print(f'Tamaño final de la población al final del período de simulación: {tamaño_final}')

# Imprimir la matriz de datos con formato legible
np.set_printoptions(suppress=True)
print("Matriz de datos (Tiempo, Población):")
print(datos)


```
